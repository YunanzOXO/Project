```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
netflix = pd.read_csv('Netflix.csv')
```

```python
# Convert date_added column to datetime object
netflix['date_added'] = pd.to_datetime(netflix['date_added'])
```

```python
# Line plots of number of titles added over time for different countries
countries = ['United States', 'India', 'United Kingdom', 'Canada']
for country in countries:
    country_netflix = netflix[netflix['country'] == country]
    title_counts_by_date = country_netflix.groupby('date_added')['show_id'].count().reset_index()
    plt.plot(title_counts_by_date['date_added'], title_counts_by_date['show_id'], label=country)
plt.legend()
plt.title('Number of Titles Added Over Time (Top 4 Countries)')
plt.xlabel('Date Added')
plt.ylabel('Number of Titles')
plt.show()
```

```python
# Pie charts of the distribution of TV show and movie ratings
ratings = ['TV-MA', 'TV-14', 'TV-PG', 'R']
for rating in ratings:
    rating_netflix = netflix[netflix['rating'] == rating]
    type_counts = rating_netflix['type'].value_counts()
    plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%')
    plt.title(f'Distribution of Titles by Rating ({rating})')
    plt.show()
```
